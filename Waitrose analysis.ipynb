## Data Cleaning ##

```python
def clean_item_price(price:str):

    if "each est." in price:
        price=price.replace("each est.","")
    if type(price)!=str:
        pass
    elif "£" in price:
        price=price.replace("£","")
    elif "p" in price:
        price=price.replace("p","")
        price="0."+price
    if "-"in price:
        price=price.split("-")[0]
    return float(price)
```

```python
def clean_price(price:str):

    if "each est." in price:
        price=price.replace("each est.","")
    if type(price)!=str:
        pass
    elif "£" in price:
        price=price.replace("£","")
    elif "p" in price:
        price=price.replace("p","")
        price="0."+price
    if "-"in price:
        price=price.split("-")[0]
    return float(price)
```

```python
mid_df['item-price'] = mid_df['item-price'].fillna('0')
mid_df['price'] = mid_df['item-price'].apply(clean_item_price)
```

## specific Product ##

```python
chees = mid_df[
    mid_df['name'].str.contains('Cheesecakes', case=False, na=False) & 
     (mid_df['category'].str.contains('everyday', case=False, na=False) | 
    ~mid_df['category'].str.contains('everyday', case=False, na=False))
]
```

```python
def process_size(size):
    if '2x' in size:
        size = size.replace('2x', '').replace('g', '')
        return float(size) * 2
    elif 'g' in size:
        return float(size.replace('g', ''))
chees['size_processed'] = chees['size'].apply(process_size)
chees['new_price'] = chees['price'] / chees['size_processed'] * 1000

chees_plot= chees[['name', 'price', 'size', 'new_price']].reset_index(drop=True)
```

```python
chees_plot
```

```python
group_chees = chees_plot.groupby('name')['new_price'].mean().sort_values(ascending=False).to_frame().reset_index()
group_chees
```

```python
(
    ggplot(group_chees, aes(x='name', y='new_price')) +
    geom_line(color='red', size=1.2) +
    labs(
        title='Price of Cheesecakes',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)
)
```
